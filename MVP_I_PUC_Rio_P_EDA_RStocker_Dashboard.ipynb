Resumo do Dataset: 

A função resumo fornece uma tabela consolidada de alto nível que usa alguns métodos disponíveis no pandas retornando ao quadro geral com os resultados organizados em colunas específicas para cada método aplicado e com os atributos empilhados em linhas. 

A primeira informação retorna como resultado da aplicação do método shape. Em uma dupla com o número de entradas (linhas e colunas) do conjunto de dados, seguido por outra coluna que retorna os tipos de dados para cada atributo, denominado "Dtype”, dessa vez retornando o resultado com o método "dtypes", outra coluna Uniques que nos retorna os valores exclusivos ao usar o método nunique, bem como a coluna Missing e % Missing para os valores ausentes presentes em cada atributo com o método isnull e finalmente uma visualização dos três primeiros e últimos registros ou entradas do conjunto de dados.

```python
def resumo ( cardio_br_IoT):
    

    # Print o Titulo com o nome da tabela enumerada
    i=1 # Numero da Tabela
    print(f'Tabela {i} Resumo do dataframe')

    # Print o tamanho do dataFrame 
    nRow, nCol = cardio_br_IoT.shape
    print(f'O conjunto de dados tem {nRow} entradas e {nCol} colunas')

    # Povoamos a Tabela

    resumo = pd.DataFrame(cardio_br_IoT.dtypes, columns=['dtypes'])
    resumo = resumo.reset_index()
    resumo['Name'] = resumo['index']
    resumo = resumo[['Name','dtypes']]
    resumo['Missing'] = cardio_br_IoT.isnull().sum().values
    resumo['% Missing']=  cardio_br_IoT.isnull().sum().values / len(cardio_br_IoT)*100     
    resumo['Uniques'] = cardio_br_IoT.nunique().values
    resumo['First Value'] = cardio_br_IoT.loc[0].values
    resumo['Second Value'] = cardio_br_IoT.loc[1].values
    resumo['Third Value'] = cardio_br_IoT.loc[2].values
    resumo['last_one'] = cardio_br_IoT.loc[cardio_br_IoT.index[-1]].values
    resumo['penultimate'] = cardio_br_IoT.loc[cardio_br_IoT.index[-2]].values
    resumo['antepenultimate'] = cardio_br_IoT.loc[cardio_br_IoT.index[-3]].values
  
    return resumo.round(0)
```

O primeiro passo é entender a composição do conjunto de dados.

Geralmente pode-se classificar os dados em uma dessas categorias: Numérico, Categórico, Texto, Data.

Na célula seguinte aprofundaremos a análise das categorias de cada atributo. Onde iremos classificar corretamente cada um dos atributos.


```python
# A função get_var_category determina a categoria de cada coluna (atributo) para o dataset carregado.

def get_var_category(series):
    unique_count = series.nunique(dropna=False)
    total_count = len(series)
    if pd.api.types.is_numeric_dtype(series):
        return 'Numérica'
    elif pd.api.types.is_datetime64_dtype(series):
        return 'Data'
    elif unique_count==total_count:
        return 'Texto (Unique)'
    else:
        return 'Categorica'

def print_categories(cardio_br_IoT):  
    for column_name in cardio_br_IoT.columns:
        print(column_name, ": ", get_var_category(cardio_br_IoT[column_name]))
```

```python
# O comando seguinte retorna a categoria de cada variável:

print_categories(cardio_br_IoT)
```

No método Winsorization se um valor exceder o valor do percentual 99º  e  embaixo do percentual  1º os valores que excedem a regra são tratados como outliers.

Com os resultados dessa análise iremos entender a natureza que o geral e a seguir como é lidar com eles ou trata-los ou removê-los.

```python
# Os outlier pode ser de dois tipos: univariado e multivariado. Acima, discutimos o exemplo de outlier univariado.
# Esses outliers podem ser encontrados quando olhamos para a distribuição de uma única variável no caso desta célula e "Distance (km)" 

# Aqui deduziremos se efetivamente eles são outliers ao conferir com outros atributos.
#q = cardio_br_IoT['Distance (km)'].quantile(0.99)

out_dist=[]

def Winsorization_outliers(df):
    q3 = np.percentile(df , 99)
    for i in df:
       if i > q3:
         out_dist.append(i)

    print(" Os supostos Outliers univariados :", out_dist)

# Winsorization_outliers atributo Distance (km)
Winsorization_outliers(cardio_br_IoT['Distance (km)'])

# Obtém os Index Supostos Outliers Univariados:
index_dist_outliers= cardio_br_IoT.index[cardio_br_IoT['Distance (km)'] >= min(out_dist)].tolist()
index_dist_outliers
#index_dist_outliers = [12, 87, 219, 246, 257]

# Análise Outliers multivariados em relaçao ao atributo Distance (Km)
col_n_dimentional_outliers = ['Type', 'Distance (km)', 'Duration', 'Average Pace', 'Average Speed (km/h)', 'Calories Burned', 'Climb (m)']
cardio_br_IoT.loc[index_dist_outliers, col_n_dimentional_outliers ]
 
# Outliers multivariados são outliers em um espaço n-dimensional.
# Aqui deduziremos se efetivamente eles são outliers e conferir com outros atributos numéricos ou n_dimençoes.
```
